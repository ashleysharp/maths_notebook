```python
import numpy as np
import sympy as smp
```

## Single variable

### Derivative formula

```python
f = lambda x: x**2

def diff(a, n):
    delta_x = 1/n
    delta_y = f(a + delta_x) - f(a)
    result = delta_y/delta_x
    return result

diff(3, 100)
```

## Symbolic derivatives

```python
x = smp.Symbol('x')
f = smp.Function('f')
g = smp.Function('g')(x)

expr = x ** 2
expr.subs(x, 3)

f = smp.Lambda(x, x**2)

expr_derivative = smp.diff(expr, x)
expr_derivative.subs(x, f)
```

## Riemann integrals

```python
def f(x):
    return x**2

def dx(a, b, n):
    return (b-a)/n

def riemann_left(a, b, n):
    delta_x = dx(a, b, n)
    total = 0
    for i in range(n):
        total += f(a + i*delta_x)*delta_x
    return total

def riemann_right(a, b, n):
    delta_x = dx(a, b, n)
    total = 0
    for i in range(1, n+1):
        total += f(a + i*delta_x)*delta_x
    return total

riemann_left(3, 5, 99)
```

## Symbolic integrals

```python
x = smp.Symbol('x')
expr = x ** 2

smp.integrate(expr, x)
smp.integrate(expr, (x, 0, 3))
```
